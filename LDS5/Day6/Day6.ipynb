```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

data = {'text_clean': ['kẻ cắp gặp bà già','bà già đụng ăn cướp','thiên linh linh địa linh linh'],
        }

speech_df = pd.DataFrame(data, columns = ['text_clean'])

cv = CountVectorizer(min_df=0.2, max_df=0.8, ngram_range = (3,3))

cv_trainsformed = cv.fit_transform(speech_df['text_clean'])
cv_array = cv_trainsformed.toarray()

cv_df = pd.DataFrame(cv_array, columns=cv.get_feature_names())
cv_df
```

```python
final_df = pd.concat([speech_df, cv_df], axis =1, sort=False)
final_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features=100)
tv.fit_transform(speech_df['text_clean'])

tv_transformed = tv.transform(speech_df['text_clean'])
tv_array = tv_transformed.toarray()
tv_df_final = pd.DataFrame(tv_array, columns=tv.get_feature_names())
tv_df_final = pd.concat([speech_df, tv_df_final], axis =1, sort=False)
tv_df_final
```

```python
df_speech = pd.read_csv("data/inaug_speeches.csv",encoding='latin1')
```

```python
df_speech
```

```python
df_speech['text_clean'] = df_speech['text'].str.lower()
df_speech['text_clean'] = df_speech['text'].str.replace(u'\xa0', u' ')
df_speech['text_clean'] = df_speech['text'].str.replace('[^a-zA-Z]', ' ')
df_speech['text_clean'] = df_speech['text'].str.replace('\s+', ' ')
```

```python
def replace_string(x):
    x = x.replace("<U+0092>", "'")
    x = x.replace("<U+0097>", " ")
    x = x.replace('[^a-zA-Z]', ' ')
    return x
df_speech['text_clean'] = df_speech['text_clean'].apply(replace_string)
```

```python
df_speech['text_clean'][57]
```

```python
cv = CountVectorizer(min_df=0.2, max_df=0.8, ngram_range = (3,3))

cv_trainsformed = cv.fit_transform(df_speech['text_clean'])
cv_array = cv_trainsformed.toarray()

cv_df = pd.DataFrame(cv_array, columns=cv.get_feature_names())
print(cv_df.shape)
print(cv.get_feature_names())

cv_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(min_df=0.2, max_df=0.8, ngram_range = (3,3))
tv.fit_transform(df_speech['text_clean'])

tv_transformed = tv.transform(df_speech['text_clean'])
tv_array = tv_transformed.toarray()
tv_df_final = pd.DataFrame(tv_array, columns=tv.get_feature_names())
tv_df_final = pd.concat([df_speech, tv_df_final], axis =1, sort=False)
print(tv_df_final.shape)
print(tv.get_feature_names())
tv_df_final
```
