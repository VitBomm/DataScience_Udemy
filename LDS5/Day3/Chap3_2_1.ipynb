# Câu 2. Billboard Top 100
## Cho dữ liệu billboard.csv. Bộ dữ liệu này đại diện cho thứ hạng hàng tuần của các bài hát kể từ thời điểm chúng lọt vào Top 100 của Billboard cho đến 75 tuần tiếp theo

* Đọc dữ liệu
* Xem xét vấn đề về dữ liệu cần khắc phục
* Chuẩn lại dữ liệu để khắc phục vấn đề trên
* Sau khi chuẩn lại dữ liệu, có điều gì phát sinh cần khắc phục tiếp theo không? Nếu có thì hãy chuẩn lại dữ liệu mới có ở câu trên.

```python
df_2 = pd.read_csv("data/billboard.csv", encoding="latin")
```

```python
df_2.head()
```

Nếu một bài hát nằm trong Top 100 dưới 75 tuần. Các cột còn lại chứa đầy cái cột thiếu (NaN)

```python
df2_after = pd.melt(frame=df_2, id_vars=["year", "artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"],
                    var_name="week", value_name="rank")
df2_after.head(10)
```

* Chuẩn hóa dữ liệu
* Thay chuỗi trong tuần bằng số
* Xóa bỏ các dòng dữ liệu Nan
    * Xem kiểu của rank, nếu chưa là số thì chuyên thành số

```python
df2_after['week'] = df2_after['week'].str.extract("(\d+)", expand=False).astype(int)
```

```python
df2_after
```

```python
df2_after = df2_after.dropna()
```

```python
df2_after.shape
```

```python
df2_after["rank"] = df2_after["rank"].astype(int)
```

#### Vấn đề mới phát sinh:
* Nhiều đơn vị quan sát(song & rank của nó ) trong một bảng duy nhất

```python
df2_after.head()
```

```python
songs_cols = ["year", "artist.inverted", "track", "time", "genre"]
songs = df2_after[songs_cols].drop_duplicates()
songs = songs.reset_index(drop=True)
songs["song_id"] = songs.index
songs.head()
```

```python
ranks = pd.merge(df2_after, songs, on=["year", "artist.inverted", "track", "time", "genre"])
ranks = ranks[["song_id", "date.entered", "date.peaked", "week", "rank"]]
ranks.head()
```

```python
# Tạo cột date
ranks["date"] = pd.to_datetime(ranks["date.entered"]) + pd.to_timedelta(ranks["week"], unit="w") - pd.DateOffset(weeks=1)
```

```python
ranks
```

```python
ranks = ranks.drop(["date.entered", "date.peaked"], axis = 1)
```
