## Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo country, year
* Hãy chuyển dữ liệu mới làm ở trên vè dạng thống kê

```python
data = pd.read_excel("data/BMI.xlsx")
```

```python
data
```

```python
df_melted = pd.melt(data, id_vars=["Country"], var_name="year", value_name="bmi")
```

```python
df_melted
```

```python
df_melted['year'] = df_melted['year'].map(lambda x: int(x[1:]))
```

```python
df_melted
```

```python
df_pivot = df_melted.pivot(index="Country", columns="year", values="bmi")
```
